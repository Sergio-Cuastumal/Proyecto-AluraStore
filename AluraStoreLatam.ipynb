## 2. Productos más y menos vendidos

Luego de conocer los ingresos totales de cada tienda quisimos saber cuáles eran los productos más y menos populares de cada tienda, esto nos ayuda a saber qué producto destaca más que otro en las respectivas tiendas, o si algún producto que se venda en una tienda no se esté vendiendo en otra. Este análisis puede ser de vital importancia ya que impacta directamente los ingresos de cada tienda.

```python


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

# 2. Graficamos cada tienda en su respectivo 'espacio' (ax)
# Tienda 1 (Arriba a la izquierda)
tops.plot(kind='barh', ax=axes[0, 0], color=dif)
axes[0, 0].set_title('Rendimiento Tienda 1')

# Tienda 2 (Arriba a la derecha)
tops2.plot(kind='barh', ax=axes[0, 1], color=dif2)
axes[0, 1].set_title('Rendimiento Tienda 2')

# Tienda 3 (Abajo a la izquierda)
tops3.plot(kind='barh', ax=axes[1, 0], color=dif3)
axes[1, 0].set_title('Rendimiento Tienda 3')

# Tienda 4 (Abajo a la derecha)
tops4.plot(kind='barh', ax=axes[1, 1], color=dif4)
axes[1, 1].set_title('Rendimiento Tienda 4')
```

Para conocer los productos más y menos vendidos se realizó un promedio de las ventas totales para cada tienda, con ello procedimos a separarlos en dos grupos: los que superan o mantienen el promedio y los que sus ventas están por debajo de él. Con ello graficamos los dos grupos solamente escogiendo los 10 productos que más se destacan en sus grupos, por lo que tenemos un top 10 de productos más y menos vendidos para cada tienda.

De esta grafica podemos identificar que en la tienda 4 no solo no se vende uno de los mejores productos de la tienda 1 como lo es el **TV LED UHD 4K**, si no que uno de los mejores productos que si comparten ambas tiendas está en lugares opuestos de sus respectivos grupos, en la tienda 1 el producto **Armario** es practicamente uno de los más vendidos y en la tienda 4 está muy cerca de ser su producto menos vendido. También podemos destacar que la tienda 3 tiene un número de ventas menor con respecto a las demás.


## 3. Calificaciones de cada tienda

Conocer la opinión de los clientes con respecto a los productos de cada tienda es importante ya que puede indicarnos cuál de todas es la favorita, esto puede ayudarnos a que la tienda genere más ventas según los productos favoritos de los clientes. Por ello realizamos dicho análisis para saber qué tan a gusto están los clientes con cada una de las tiendas.

```python
# Graficamos con puntos y una línea tenue
calificaciones = pd.Series(calificaciones)
calificaciones.plot(kind='line', marker='D', markersize=12, linewidth=1, color='gold', markerfacecolor='orange')

# EL TRUCO: Ajustar el límite del eje Y para resaltar la diferencia
plt.ylim(calificaciones.min() - 0.05, calificaciones.max() + 0.05)

plt.title('Comparativa Detallada de Calificaciones')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
```

Como las calificaciones de las tiendas es bastante similar se optó por un grafico de eje recortado que se enfoca en la parte superior de sus elementos a comparar. En el grafico podemos ver que la mejor tienda es la tienda 3, seguida de la tienda 2. La tienda 1 es la peor en valoración de las cuatro tiendas seguida de la tienda 4. Esta ultima comparación es importante ya que una de las tiendas que más ingresos tiene es la que también tiene peor calificación, y la tienda que menos ingresos tiene es una de las que peor calificación tiene también. Esto nos dice que la tienda 4 no solo es la que menos ingresos genera si no que también es una de las peores valoradas por los clientes.

## 4. Categorias más vendidas por tienda

Las ventas por categoría también son importantes ya que la diferencia entre las ventas puede deberse a que no se están vendiendo productos de una categoría en una tienda que en otra si se venden más. Dicho analisis se realiza en cada para saber la distribución de las ventas de cada categoría.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

# 2. Graficamos cada tienda en su respectivo 'espacio' (ax)
# Tienda 1 (Arriba a la izquierda)


ventasxCategoria.plot(kind='pie', ax=axes[0, 0] ,autopct='%1.1f%%')
axes[0, 0].set_title('Productos más vendidos de la Tienda 1')
ventasxCategoria2.plot(kind='pie', ax=axes[0, 1] ,autopct='%1.1f%%')
axes[0, 1].set_title('Productos más vendidos de la Tienda 2')
ventasxCategoria3.plot(kind='pie', ax=axes[1, 0] ,autopct='%1.1f%%')
axes[1, 0].set_title('Productos más vendidos de la Tienda 3')
ventasxCategoria4.plot(kind='pie', ax=axes[1, 1] ,autopct='%1.1f%%')
axes[1, 1].set_title('Productos más vendidos de la Tienda 4')
```
